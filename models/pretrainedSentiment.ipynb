```python
import pandas as pd
from germansentiment import SentimentModel

model = SentimentModel()

# Load your data
data = pd.read_excel('subset_studium_german.xlsx')

#columns = ['Vorbereitung', 'Unterkunft', 'Studium', 'Freizeit', 'Budget', 'Fazit']
#Fazit rather positiv, Unterkunft to specific to the dorm, Freizeit rather positiv, Budget to many numeric only values
columns = ['Studium']
for x in columns:
    data[x] = data[x].astype(str)
    results = model.predict_sentiment(data[x].iloc[:500])
    print("------------------------------")
    print("Results of: " + x)
    print("negative entries: " + str(results.count('negative')))
    print("positive entries: " + str(results.count('positive')))
    print("neutral entries: " + str(results.count('neutral')))
    print("------------------------------")
    studium_text = pd.DataFrame(data[x].iloc[:500])
    studium_text.insert(1, "Sentiment", results, True)
    studium_text.to_excel(x+".xlsx")
# Function to apply sentiment analysis
```
