```python
import pandas as pd
from langdetect import detect
```

```python
df = pd.read_csv(r"./data/final_labeling_WIP.csv", sep=";")
df.set_index("Unnamed: 0", inplace=True)
```

```python
df.insert(30, "Sprache", "")
```

```python
for index, row in df.iterrows():
      df.at[index, 'Sprache'] = detect(str(row["Studium"]))
```

```python
df["Sprache"].value_counts()
```

```python
df_de = df[df['Sprache'] == "de"]
```

```python
df_en = df[df['Sprache'] == "en"]
```

```python
# the single value in df_tl is wrongly labeled and is indeed a german entry
df_tl = df[df["Sprache"] == "tl"]
```

```python
df_de = pd.concat([df_de, df_tl])
```

```python
df_de.to_csv("all_data_german.csv", sep=";")
```

```python
df_en.to_csv("all_data_english.csv", sep=";")
```
